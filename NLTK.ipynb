词干提取

```python
print(text_2)
```

```python
from nltk.stem import LancasterStemmer, PorterStemmer, SnowballStemmer
from nltk.tokenize import word_tokenize

# Porter词干提取法
stemmer = PorterStemmer()
token_content=word_tokenize(text_2)
stemmed_words = [stemmer.stem(i) for i in token_content]
print('Porter Stemmer')
print(stemmed_words)

# Snowball词干提取法
stemmer = SnowballStemmer('english')
token_content=word_tokenize(text_2)
stemmed_words = [stemmer.stem(i) for i in token_content]
print('Snowball Stemmer')
print(stemmed_words)

# Lancaster词干提取法
stemmer = LancasterStemmer()
token_content=word_tokenize(text_2)
stemmed_words = [stemmer.stem(i) for i in token_content]
print('Lancaster Stemmer')
print(stemmed_words)
```

# 统计自然语言处理

NER

```python
import nltk

loc=[('TCS', 'IN', 'PUNE'),
    ('INFOCEPT', 'IN', 'PUNE'),
    ('WIPRO', 'IN', 'PUNE'),
    ('AMAZON', 'IN', 'HYDERABAD'),
    ('INTEL', 'IN', 'HYDERABAD'),
    ]

query = [e1 for (e1, rel, e2) in loc if e2=='HYDERABAD']
print(query)
query = [e1 for (e1, rel, e2) in loc if e2=='PUNE']
print(query)
```

分块（chunk）

```python
sentence = [("the", "DT"), ("little", "JJ"), ("yellow", "JJ"), ("dog", "NN"), ("barked", "VBD"), ("at", "IN"),  ("the", "DT"), ("cat", "NN")]
grammar = "NP: {<DT>?<JJ>*<NN>}"
cp = nltk.RegexpParser(grammar)
result = cp.parse(sentence)
print(result)
```

正则表达式分块

```python
import nltk
grammar = r"""  NP: {<DT|PP\$>?<JJ>*<NN>}
            {<NNP>+}
            """
cp = nltk.RegexpParser(grammar)
sentence = [("Rapunzel", "NNP"), ("let", "VBD"), ("down", "RP"), ("her", "PP$"), ("long", "JJ"), ("golden", "JJ"), ("hair", "NN")]
print(cp.parse(sentence))
```

机器翻译

```python
from nltk.translate.ibm1 import IBMModel1
from nltk.translate import AlignedSent

bitext = []
bitext.append(AlignedSent(['klein', 'ist', 'das', 'haus'], ['the', 'house', 'is', 'small']))
bitext.append(AlignedSent(['das', 'haus', 'ist', 'ja', 'groß'], ['the', 'house', 'is', 'big']))
bitext.append(AlignedSent(['das', 'buch', 'ist', 'ja', 'klein'], ['the', 'book', 'is', 'small']))
bitext.append(AlignedSent(['das', 'haus'], ['the', 'house']))
bitext.append(AlignedSent(['das', 'buch'], ['the', 'book']))
bitext.append(AlignedSent(['ein', 'buch'], ['a', 'book']))
ibm1 = IBMModel1(bitext, 5)
print(round(ibm1.translation_table['buch']['book'], 3))
print(round(ibm1.translation_table['das']['book'], 3))
print(round(ibm1.translation_table['buch'][None], 3))
print(round(ibm1.translation_table['ja'][None], 3))
test_sentence = bitext[2]
print(test_sentence.words)
print(test_sentence.mots)
print(test_sentence.alignment)
```

BLEU Score

```python
from nltk.translate.bleu_score import SmoothingFunction, sentence_bleu
hypothesis1 = ['It', 'is', 'a', 'guide', 'to', 'action', 'which','ensures', 'that', 'the', 'military', 'always','obeys', 'the', 'commands', 'of', 'the', 'party']
hypothesis2 = ['It', 'is', 'to', 'insure', 'the', 'troops','forever', 'hearing', 'the', 'activity', 'guidebook','that', 'party', 'direct']
reference1 = ['It', 'is', 'a', 'guide', 'to', 'action', 'that','ensures', 'that', 'the', 'military', 'will', 'forever','heed', 'Party', 'commands']
reference2 = ['It', 'is', 'the', 'guiding', 'principle', 'which','guarantees', 'the', 'military', 'forces', 'always','being', 'under', 'the', 'command', 'of', 'the','Party']
reference3 = ['It', 'is', 'the', 'practical', 'guide', 'for', 'the','army', 'always', 'to', 'heed', 'the', 'directions','of', 'the', 'party']

print(sentence_bleu([reference1, reference2, reference3], hypothesis1))
print(round(sentence_bleu([reference1, reference2, reference3], hypothesis2),4) )

chencherry = SmoothingFunction()
sentence_bleu([reference1, reference2, reference3], hypothesis2, smoothing_function=chencherry.method1) # doctest: +ELLIPSIS
```
